```python
from sklearn.preprocessing import MinMaxScaler
sc_x=MinMaxScaler()
sc_x.fit(X_train)
```

```python
X_train=pd.DataFrame(data=sc_x.transform(X_train),
                    columns=X_train.columns,
                    index=X_train.index)
```

```python
X_test=pd.DataFrame(data=sc_x.transform(X_test),
                    columns=X_test.columns,
                    index=X_test.index)
```
